### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 11)         # x values from 0 to 10
y = x ** 2                   # y values as square of x

# Plot
plt.plot(x, y, marker='o')   # Line plot with markers

# Customization
plt.title('Square of Numbers from 0 to 10')
plt.xlabel('X Values')
plt.ylabel('Y = X Squared')
plt.grid(True)

# Show plot
plt.show()
```
