### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Create the figure and 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# 1. Line chart in the first subplot
x = list(range(0, 11))
y = [i**2 for i in x]
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y = X^2')

# 2. Bar chart in the second subplot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axs[0, 1].bar(categories, values, color=['red', 'green', 'blue', 'orange'])
axs[0, 1].set_title('Bar Chart')

# 3. Scatter plot in the third subplot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
axs[1, 0].set_title('Scatter Plot')
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f'({x_scatter[i]},{y_scatter[i]})', (x_scatter[i], y_scatter[i]))

# 4. Pie chart in the fourth subplot
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  # Equal aspect ratio for a circle

# Adjust layout and display
plt.tight_layout()
plt.show()
```
