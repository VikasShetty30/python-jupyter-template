### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'orange']  # Different colors for each bar

# Plot
plt.bar(categories, values, color=colors)

# Customization
plt.title('Category-wise Values')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional: adds a light horizontal grid

# Show plot
plt.show()
```
